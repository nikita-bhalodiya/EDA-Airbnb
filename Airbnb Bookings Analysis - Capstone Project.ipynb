## 1) Analyse data with respect to room_category Vs Price

Let's check the relation between neighbourhood_group and room type with Price.

```python
# Analyse data with respect to room_category
mean_price_df = df.groupby(['neighbourhood_group', 'room_type'], as_index=False)[['price']].mean()
sns.barplot(data=mean_price_df, x='neighbourhood_group', y ='price', hue='room_type')
```

Some Conclusion from above graph:

1) We can say that Manhattan have highest price booking and second highest price booking is Brooklyn followed by Staten Island.


2) In Some area like Bronx,Queens and Staten Island have almost same price for both private and share room.

3) Most active and urban area like Manhattan and Brooklyn have price different between private and shared room.


```python

fig = plt.figure(figsize=(24, 6))
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)
sns.lineplot(data=df, x='neighbourhood', y='price', hue='room_type')
plt.title('Distribution of Prices in neighbourhoods')
sns.despine(fig)
```

All three type of room highest price is for Entire_home or apt for almost all neighbourhoods.

```python
# Average Price for each room_category

mean_price_df = df.groupby('room_type', as_index=False)[['price']].mean()
sns.barplot(data=mean_price_df, x='room_type', y='price', palette='GnBu_d').set_title("Avg Price vs Rooms type")
```

Entire home/apt have highest price and Shared room have lowest price.

## 2) Analyse data with respect to Avg Minimum Nights vs Neighbourhood / room_category


```python
# how many nights people stayed in each room_category
min_night_df = df.groupby('room_type', as_index=False)[['minimum_nights']].mean()
sns.barplot(data=min_night_df, x='room_type', y='minimum_nights', palette='GnBu_d').set_title("Minimum Nights vs Rooms type")
```

Entire home/apt have highest avg minimun night stay restriction.

---



```python
# Avg Minimum Nights vs Neighbourhood / Room Type

mean_price_df = df.groupby(['neighbourhood_group', 'room_type'], as_index=False)[['minimum_nights']].mean()
sns.barplot(data=mean_price_df, x='neighbourhood_group', y='minimum_nights', hue='room_type', palette='GnBu_d').set_title("Avg Minimum Nights vs Neighbourhood / Room Type")
```

We can conclude that Manhattan has highest avg minimum night restriction.

```python
# Price vs Minimum Nights
sns.scatterplot(data=df[df['price'] < 1000], x='minimum_nights', y='price', hue='room_type', palette="ch:r=-.2,d=.3_r").set_title("Price vs Minimum Nights (Range $ 0 - 1000)")
```

the more cost the property has, the more it tends to allow lower number of minimum nights

## 3)Analyse data with respect monthly reviews varies with room_category in each neighbourhood groups

```python
# monthly reviews varies with room_category in each neighbourhood groups

sns.lineplot(data=df, x='neighbourhood_group', y='reviews_per_month', hue='room_type', style="room_type",
        markers=True, dashes=False).set_title("Review per Month vs Neighbourhoods")
```

Brooklyn and Manhattan stands pretty fall in terms of review rate per month.

```python
# Price vs Number of Reviews
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='room_type', palette="ch:r=-.2,d=.3_r",
        hue_order=['Private room', 'Shared room', 'Entire home/apt']).set_title('Price vs Number of Reviews')
```

we can notice a negative relation between price & no. of reviews. Where, costlier properties has significantly less no of reviews, but cheaper properties have large number of reviews. Usually, cheaper rooms has more number of guest visits than costlier one, we know no of reviews is directly proportional to no of guests.

## 4) Analyse data with respect to room availability with neighbourhoods, room standards

```python
sns.lineplot(data=df, x='neighbourhood_group', y='availability_365', hue='room_type').set_title('Room Availability throughout Neighbourhood/Room Type')
```

Staten Island has most busiest Shared Room as well as Most available Private Room. Brroklyn and Manhattan has quite similar availabilities

```python
sns.scatterplot(data=df[df['price'] < 500], x="availability_365", y='price', hue='room_type', alpha=.9, palette="muted").set_title('Price vs Availability (Range $ 0 - 500)')
```

we can notice almost uniform distribution i.e. A particular room type of a particular price range can be available throughout year.

```python
sns.countplot(data=df[df['availability_365']  == 365], x='neighbourhood_group', hue='room_type', palette='GnBu_d').set_title('Property Available 365 days')
```

Also there are 1295 total properties that are available 365 days a year. Manhattan tops this list with 572 properties while Brooklyn stands at second with 453 properties available.
